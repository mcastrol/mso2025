1) Definir una función python llamada "simular_ruleta" que reproduzca n tiradas de una ruleta con numeros de 0 a 36.


```python
import random
import matplotlib.pyplot as plt
from collections import Counter

def simular_ruleta(n):
    """
    Simulate n spins of a roulette wheel (numbers 0-36) and plot a histogram.
    Also displays the numbers that appeared most and least frequently.

    Parameters:
    n (int): Number of spins to simulate

    Returns:
    list: List of spin results
    dict: Frequency count of each number
    """
    # European roulette has numbers 0-36
    spins = [random.randint(0, 36) for _ in range(n)]
    freq = Counter(spins)

    # Create histogram
    plt.figure(figsize=(12, 6))
    plt.hist(spins, bins=37, range=(-0.5, 36.5), density=True, edgecolor='black', alpha=0.7)
    plt.title(f'Roulette Spin Distribution (n={n:,} spins)')
    plt.xlabel('Number')
    plt.ylabel('Relative Frequency')
    plt.xticks(range(0, 37))
    plt.grid(axis='y', alpha=0.5)

    # Calculate and plot expected uniform distribution
    expected = 1/37
    plt.axhline(y=expected, color='red', linestyle='--', label='Expected Frequency (1/37)')
    plt.legend()

    # Find most and least frequent numbers
    most_common = freq.most_common(1)[0]
    least_common = freq.most_common()[-1]

    # Add text box with frequency information
    info_text = (f"Most frequent: Number {most_common[0]} (appeared {most_common[1]} times, {most_common[1]/n:.2%})\n"
                 f"Least frequent: Number {least_common[0]} (appeared {least_common[1]} times, {least_common[1]/n:.2%})")

    plt.gcf().text(0.15, 0.85, info_text, bbox=dict(facecolor='white', alpha=0.8),
                fontsize=10, verticalalignment='top')

    plt.show()

    # Print frequency information to console
    print(f"\nResults for {n:,} spins:")
    print(f"Most frequent number: {most_common[0]} (appeared {most_common[1]} times, {most_common[1]/n:.2%})")
    print(f"Least frequent number: {least_common[0]} (appeared {least_common[1]} times, {least_common[1]/n:.2%})")

    return spins, freq

# Run simulations with different sample sizes
sample_sizes = [100, 500, 1000, 5000, 10000]
results = {}

for size in sample_sizes:
    spins, freq = simular_ruleta(size)
    results[size] = {'spins': spins, 'frequency': freq}
    print("-" * 60)
```

## **EJERCICIO 4**:
Definir una función de devuelva n valores de una distribución poisson de lambda l.
Definir una función que imprima el histograma. Probar ambas funciones para varios tamaños y madia lambda=3 clientes / unidad de tiempo.

```python
# DEFINICIÓN FUNCIONES
def randomPoisson(l, s):
  x=np.random.poisson(lam=l, size=s)
  return(x)

def distribucionPoisson(l, s):
  x = randomPoisson(l, s)
  print(x)
  print(f"-- HISTOGRAMA: lamda={l}, size={s}")
  plt.figure()  # Crea un figura para cada histograma
  plt.hist(x,bins=15)
  plt.show()

# Lambda = 1 , Size = 100
distribucionPoisson(1, 100)

# Lambda = 1 , Size = 500
distribucionPoisson(1, 500)

# Lambda = 3 , Size = 100
distribucionPoisson(3, 100)

# Lambda = 3 , Size = 500
distribucionPoisson(3, 500)

```

## **EJERCICIO 5:**

Definir una función de devuelva n valores de una distribución exponencial de scale 1/lambda.
Definir una función que imprima el histograma. Probar ambas funciones para varios tamaños y lambda=3.
Imprimir los valores máximos y mínimos obtenidos

```python
# DEFINICIÓN FUNCIONES
def randomExponencial(l, s):
  x=np.random.exponential(scale=1/l, size=100)
  return(x)

def minValue(x):
  return x.min()

def maxValue(x):
  return x.max()

def distribucionExponencial(l, s):
  x = randomExponencial(l, s)
  print(x)
  print(f"-- HISTOGRAMA: scale=1/{l}, size={s}")
  plt.figure()  # Crea un figura para cada histograma
  plt.hist(x)
  print("MIN: ", minValue(x))
  print("MAX: ", maxValue(x))
  plt.show()


# Scale = 1/3 , Size = 100
distribucionExponencial(3, 100)

# Scale = 1/3 , Size = 5000
distribucionExponencial(3, 5000)
```
